```python
import os
import pandas as pd

IS_KAGGLE = os.path.exists('/kaggle/input')
BASE = '/kaggle/input/competitions/home-credit-default-risk/'

if IS_KAGGLE:
    app   = pd.read_csv(BASE + 'application_train.csv')
    inst  = pd.read_csv(BASE + 'installments_payments.csv')
    bur   = pd.read_csv(BASE + 'bureau.csv')
    bur_b = pd.read_csv(BASE + 'bureau_balance.csv')
    pos   = pd.read_csv(BASE + 'POS_CASH_balance.csv')
    prev_app = pd.read_csv(BASE + 'previous_application.csv')
    cc    = pd.read_csv(BASE + 'credit_card_balance.csv')

    out = '/kaggle/working/local_smoketest/'
    os.makedirs(out, exist_ok=True)

    app_sample = app[app['TARGET'].notna()].head(2000)
    sample_ids = set(app_sample['SK_ID_CURR'].values)

    inst_sample   = inst[inst['SK_ID_CURR'].isin(sample_ids)]
    bur_sample    = bur[bur['SK_ID_CURR'].isin(sample_ids)]
    bur_ids       = set(bur_sample['SK_ID_BUREAU'].values)
    bur_b_sample  = bur_b[bur_b['SK_ID_BUREAU'].isin(bur_ids)]

    prev_app_sample = prev_app[prev_app['SK_ID_CURR'].isin(sample_ids)]
    prev_ids       = set(prev_app_sample['SK_ID_PREV'].values)
    pos_sample     = pos[pos['SK_ID_PREV'].isin(prev_ids)]
    cc_sample      = cc[cc['SK_ID_PREV'].isin(prev_ids)]

    app_sample.to_csv(out + 'application_train.csv',       index=False)
    inst_sample.to_csv(out + 'installments_payments.csv',  index=False)
    bur_sample.to_csv(out + 'bureau.csv',                  index=False)
    bur_b_sample.to_csv(out + 'bureau_balance.csv',        index=False)
    prev_app_sample.to_csv(out + 'previous_application.csv', index=False)
    pos_sample.to_csv(out + 'POS_CASH_balance.csv',        index=False)
    cc_sample.to_csv(out + 'credit_card_balance.csv',      index=False)

    for name, df in [
        ('app', app_sample), ('inst', inst_sample), ('bur', bur_sample),
        ('bur_b', bur_b_sample), ('prev_app', prev_app_sample),
        ('pos', pos_sample), ('cc', cc_sample)
    ]:
        print(f'{name}: {df.shape}')
    print(f'Saved 7 files to {out}')
    print('Copy contents to data/home_credit/ to run 00_pd_homecredit_lstm_kaggle.ipynb locally.')
```
